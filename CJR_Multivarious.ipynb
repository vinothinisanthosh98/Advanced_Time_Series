```python
"""
Advanced Time Series Forecasting Project
File: CJR_time_series_improved.py
- EDA
- Feature engineering (time features, lags, rolling)
- Scaling
- Positional encoding
- Transformer model with attention extraction
- Callbacks (EarlyStopping, ReduceLROnPlateau, ModelCheckpoint)
- Training, evaluation (MAE, RMSE), plots
- Future multi-step forecasting (autoregressive)

Usage: put your CSV as 'continuous dataset.csv' in the same folder and run.
Requires: pandas, numpy, matplotlib, seaborn, scikit-learn, tensorflow

"""

import os
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
import tensorflow as tf
from tensorflow.keras import layers, Model, callbacks, optimizers
from tensorflow.keras.optimizers import Adam

# ----------------------------
# CONFIG
# ----------------------------
DATA_PATH = "continuous dataset.csv"
SEQ_LEN = 48           # number of past steps used as input
FORECAST_HORIZON = 1   # steps ahead for training target (1 = next step)
FUTURE_STEPS = 24      # when doing autoregressive forecasting
BATCH_SIZE = 64
EPOCHS = 50
TEST_SPLIT = 0.2
VAL_SPLIT = 0.2
MODEL_SAVE = "best_transformer_ts.h5"
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ----------------------------
# 1. LOAD & BASIC CHECKS
# ----------------------------
assert os.path.exists(DATA_PATH), f"Dataset not found at {DATA_PATH}"
print("Loading dataset...", DATA_PATH)
df = pd.read_csv(DATA_PATH)
print(f"Initial shape: {df.shape}")

# infer datetime column: assume first column or named 'datetime'/'timestamp'
if 'datetime' in df.columns:
    datetime_col = 'datetime'
elif 'timestamp' in df.columns:
    datetime_col = 'timestamp'
else:
    datetime_col = df.columns[0]

# convert and set as index
df[datetime_col] = pd.to_datetime(df[datetime_col])
df = df.sort_values(datetime_col).set_index(datetime_col)
print(f"Using datetime index column: {datetime_col}")

# quick overview
print(df.head())
print(df.describe().T)

# ----------------------------
# 2. Exploratory Data Analysis (brief automated)
# ----------------------------
# missing values
missing = df.isna().sum()
print("Missing values per column:\n", missing[missing>0])

# Plot a few sample columns (first 4 numeric)
num_cols = df.select_dtypes(include='number').columns.tolist()
plot_cols = num_cols[:4]
if len(plot_cols) > 0:
    df[plot_cols].plot(subplots=True, figsize=(12, 2*len(plot_cols)), title='Sample time series')
    plt.tight_layout()
    plt.show()

# correlation heatmap (sample of numeric columns if too many)
corr_cols = num_cols if len(num_cols)<=20 else num_cols[:20]
plt.figure(figsize=(10,8))
sns.heatmap(df[corr_cols].corr(), annot=False, cmap='vlag')
plt.title('Correlation (sample)')
plt.show()

# ----------------------------
# 3. FEATURE ENGINEERING
# ----------------------------
print("Creating time features and lags...")
# time features
df_feats = df.copy()
idx = df_feats.index
# ensure frequency if possible
try:
    inferred = pd.infer_freq(idx)
    print("Inferred frequency:", inferred)
except Exception:
    inferred = None

# add calendar features
df_feats['hour'] = df_feats.index.hour
df_feats['day'] = df_feats.index.day
df_feats['month'] = df_feats.index.month
df_feats['weekday'] = df_feats.index.weekday
df_feats['is_weekend'] = (df_feats['weekday'] >= 5).astype(int)

# rolling features for first few numeric columns
for c in num_cols[:6]:
    df_feats[f'{c}_lag1'] = df_feats[c].shift(1)
    df_feats[f'{c}_roll24_mean'] = df_feats[c].rolling(window=24, min_periods=1).mean().shift(1)
    df_feats[f'{c}_roll168_mean'] = df_feats[c].rolling(window=168, min_periods=1).mean().shift(1)

# drop rows with NaNs introduced by lags
df_feats = df_feats.dropna()
print("After feature engineering shape:", df_feats.shape)

# select features (all numeric)
features = df_feats.select_dtypes(include='number').columns.tolist()
print(f"Total numeric features used: {len(features)}")

# ----------------------------
# 4. SCALING
# ----------------------------
scaler = StandardScaler()
scaled_values = scaler.fit_transform(df_feats[features])
scaled = pd.DataFrame(scaled_values, index=df_feats.index, columns=features)

# ----------------------------
# 5. SEQUENCE WINDOWING
# ----------------------------

def make_sequences(df_values, seq_len, horizon=1):
    X, y = [], []
    for i in range(len(df_values) - seq_len - (horizon-1)):
        X.append(df_values[i:i+seq_len])
        y.append(df_values[i+seq_len+horizon-1])
    return np.array(X), np.array(y)

X, y = make_sequences(scaled.values, SEQ_LEN, FORECAST_HORIZON)
print("X shape:", X.shape, "y shape:", y.shape)

# train-test split preserving time order
n = len(X)
train_n = int(n * (1 - TEST_SPLIT))
X_train, X_test = X[:train_n], X[train_n:]
y_train, y_test = y[:train_n], y[train_n:]

print(f"Train samples: {X_train.shape[0]}, Test samples: {X_test.shape[0]}")

# ----------------------------
# 6. Positional Encoding (sinusoidal)
# ----------------------------

def get_positional_encoding(seq_len, d_model):
    positions = np.arange(seq_len)[:, np.newaxis]    # (seq_len, 1)
    dims = np.arange(d_model)[np.newaxis, :]         # (1, d_model)
    angle_rates = 1 / np.power(10000, (2 * (dims//2)) / np.float32(d_model))
    angle_rads = positions * angle_rates
    pos_encoding = np.zeros((seq_len, d_model))
    pos_encoding[:, 0::2] = np.sin(angle_rads[:, 0::2])
    pos_encoding[:, 1::2] = np.cos(angle_rads[:, 1::2])
    return tf.cast(pos_encoding, dtype=tf.float32)

# ----------------------------
# 7. Transformer Model (with ability to return attention weights)
# ----------------------------

def build_transformer_model(seq_len, n_features, d_model=64, num_heads=4, ff_dim=128, dropout=0.1):
    # input
    inputs = layers.Input(shape=(seq_len, n_features), name='inputs')

    # project input to d_model
    x = layers.Dense(d_model, name='input_projection')(inputs)

    # add positional encoding
    pos_encoding = get_positional_encoding(seq_len, d_model)
    x = x + pos_encoding

    # multi-head attention block
    # return attention scores by using the functional MultiHeadAttention with return_attention_scores=True
    mha = layers.MultiHeadAttention(num_heads=num_heads, key_dim=d_model//num_heads, name='mha')
    att_output, att_scores = mha(x, x, return_attention_scores=True)
    att_output = layers.Dropout(dropout)(att_output)
    x = layers.LayerNormalization(epsilon=1e-6)(x + att_output)

    # feed-forward
    ff = layers.Dense(ff_dim, activation='relu')(x)
    ff = layers.Dense(d_model)(ff)
    x = layers.LayerNormalization(epsilon=1e-6)(x + ff)

    # pooling and output
    x = layers.GlobalAveragePooling1D()(x)
    outputs = layers.Dense(n_features, name='outputs')(x)

    model = Model(inputs=inputs, outputs=outputs, name='transformer_ts')
    model.compile(optimizer=Adam(learning_rate=1e-3), loss='mse')

    # wrap to also return attention scores in a separate model for visualization
    # create a model that maps inputs -> attention scores
    att_model = Model(inputs=inputs, outputs=att_scores, name='att_scores_model')

    return model, att_model

n_features = X.shape[2]
model, att_model = build_transformer_model(SEQ_LEN, n_features=n_features)
model.summary()

# ----------------------------
# 8. CALLBACKS
# ----------------------------
es = callbacks.EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True)
rp = callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=4, min_lr=1e-6)
mc = callbacks.ModelCheckpoint(MODEL_SAVE, monitor='val_loss', save_best_only=True)

# ----------------------------
# 9. TRAINING
# ----------------------------
history = model.fit(
    X_train, y_train,
    validation_split=VAL_SPLIT,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=[es, rp, mc],
    verbose=2
)

# save final model
model.save('final_' + MODEL_SAVE)

# ----------------------------
# 10. EVALUATION
# ----------------------------
print("Evaluating on test set...")
pred_scaled = model.predict(X_test)

# inverse scaling: scaler was fit on entire feature set; we need to invert per-sample
# Since y contains all features, we can inverse transform directly
# But ensure shapes align

y_test_inv = scaler.inverse_transform(y_test)
pred_inv = scaler.inverse_transform(pred_scaled)

mae = mean_absolute_error(y_test_inv, pred_inv)
rmse = np.sqrt(mean_squared_error(y_test_inv, pred_inv))
print(f"MAE: {mae:.4f} | RMSE: {rmse:.4f}")

# plot actual vs predicted for first few features
n_plot = min(4, n_features)
fig, axes = plt.subplots(n_plot, 1, figsize=(12, 3*n_plot))
for i in range(n_plot):
    axes[i].plot(y_test_inv[:500, i], label='actual')
    axes[i].plot(pred_inv[:500, i], label='pred')
    axes[i].set_title(features[i])
    axes[i].legend()
plt.tight_layout()
plt.show()

# error distribution
errors = (y_test_inv - pred_inv).reshape(-1)
plt.figure(figsize=(8,4))
plt.hist(errors, bins=80)
plt.title('Prediction Error Distribution (all features flattened)')
plt.show()

# ----------------------------
# 11. Attention visualization (for a single batch)
# ----------------------------
# att_model outputs attention scores of shape (batch, num_heads, seq_len, seq_len)
if hasattr(att_model, 'predict'):
    sample_input = X_test[:1]
    att_scores = att_model.predict(sample_input)
    # att_scores shape: (1, num_heads, seq_len, seq_len) or (num_heads, seq_len, seq_len) depending on TF version
    att_scores = np.squeeze(att_scores)
    if att_scores.ndim == 3:
        num_heads = att_scores.shape[0]
        plt.figure(figsize=(12, 3 * num_heads))
        for h in range(num_heads):
            plt.subplot(num_heads, 1, h+1)
            plt.imshow(att_scores[h], aspect='auto')
            plt.colorbar()
            plt.title(f'Attention head {h+1}')
        plt.tight_layout()
        plt.show()
    else:
        print('Unexpected attention score shape:', att_scores.shape)

# ----------------------------
# 12. FUTURE FORECAST (autoregressive multi-step)
# ----------------------------
print(f"Generating {FUTURE_STEPS}-step autoregressive forecast...")
# start from the last available sequence in scaled data
last_seq = scaled.values[-SEQ_LEN:].copy()  # shape (seq_len, n_features)
future_preds = []
current_seq = last_seq.copy()
for step in range(FUTURE_STEPS):
    input_seq = current_seq[np.newaxis, ...]  # (1, seq_len, n_features)
    pred_scaled_step = model.predict(input_seq)[0]
    # append prediction
    future_preds.append(pred_scaled_step)
    # roll sequence and append predicted step as new last row
    current_seq = np.vstack([current_seq[1:], pred_scaled_step])

future_preds = np.array(future_preds)  # (future_steps, n_features)
future_inv = scaler.inverse_transform(future_preds)

# create datetime index for future
freq = pd.infer_freq(df_feats.index)
if freq is None:
    # fallback: use most common difference
    diffs = np.diff(df_feats.index.values).astype('timedelta64[s]').astype(int)
    most_common = np.bincount(diffs).argmax()
    freq = f"{most_common}S"

last_time = df_feats.index[-1]
future_index = pd.date_range(start=last_time + pd.Timedelta(1, unit='h'), periods=FUTURE_STEPS, freq=freq)

future_df = pd.DataFrame(future_inv, index=future_index, columns=features)
print(future_df.head())

# plot a main target if available (assume first numeric column is main)
main_col = num_cols[0]
if main_col in future_df.columns:
    plt.figure(figsize=(12,4))
    # show last 7 days of main_col and future forecast
    historical_window = df[main_col].iloc[-24*7:]
    plt.plot(historical_window.index, historical_window.values, label='history')
    plt.plot(future_df.index, future_df[main_col].values, marker='o', label='forecast')
    plt.legend()
    plt.title(f'Historical + Forecast for {main_col}')
    plt.show()

# ----------------------------
# 13. Save artifacts & quick report
# ----------------------------
report = {
    'mae': float(mae),
    'rmse': float(rmse),
    'train_samples': int(X_train.shape[0]),
    'test_samples': int(X_test.shape[0]),
    'seq_len': SEQ_LEN,
    'n_features': n_features
}

import json
with open('report.json', 'w') as f:
    json.dump(report, f, indent=2)

print('Report saved to report.json')
print('Model saved to', 'final_' + MODEL_SAVE)

# End of file
```
